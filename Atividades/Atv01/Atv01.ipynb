Apresente a soma de todos os argumentos inteiros ou reais recebidos

```java
import java.util.Scanner;
Scanner in = new Scanner(System.in);
float num,soma = 0;

do{
    System.out.println("Digite um numero para a soma:");
    try{
        num=in.nextFloat();
    }catch(Exception e){
        break;
    }
    soma+=num;
}while(true);
System.out.println("Soma:"+soma);
```
